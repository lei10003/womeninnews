```python
def createWeightedGraph(g_test):
    G = nx.Graph()
    for u,v,data in g_test.edges(data=True):
        w = data['weight'] if 'weight' in data else 1.0
        if G.has_edge(u,v):
            G[u][v]['weight'] += int(w)
        else:
            G.add_edge(u, v, weight=w)
    return G

#G.add
```
